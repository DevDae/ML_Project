```python
import pandas as pd
from collections import Counter
from numpy.random import RandomState
from imblearn.over_sampling import SMOTENC
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_low = pd.read_csv("/content/drive/MyDrive/기계학습팀플/코드/data/train_low.csv", encoding = 'utf-8-sig')
train_high= pd.read_csv("/content/drive/MyDrive/기계학습팀플/코드/data/train_high.csv", encoding = 'utf-8-sig')
```

```python
dfTrain = pd.concat([train_low,train_high])
```

```python
dfTrain['Status'] = [1 if s >90 else 0 for s in dfTrain['처리효율']] 
```

```python
dfTrain.groupby('Status').size()
```

```python
dfX = dfTrain.drop(['Status'],axis = 1)
dfY = dfTrain['Status']
```

```python
sm = SMOTE(random_state=2022)
X_res, Y_res = sm.fit_resample(dfX, dfY)
print(f'Resampled dataset samples per class {Counter(Y_res)}')
```
