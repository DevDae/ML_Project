```python
#다중대치
# Create kernel. 
kds = mf.ImputationKernel(
  dataMF,
  datasets=4,
  save_all_iterations=True,
  random_state=1991
)

# Run the MICE algorithm for 3 iterations
kds.mice(3)

# Return the completed kernel data
completed_dataset = kds.complete_data(dataset=0, inplace=False)
print(completed_dataset.isnull().sum(0))
```

```python
import datetime
```

```python
df['가동개시일_datetime'] = pd.to_datetime(df['가동개시일'])
```

```python
df['준공일_datetime'] = pd.to_datetime(df['준공일'])
```

```python
df['가동개시일_연도'] = df['가동개시일_datetime'].dt.year
df['가동개시일_월'] = df['가동개시일_datetime'].dt.month
df['가동개시일_일'] = df['가동개시일_datetime'].dt.day
df['준공일_연도'] = df['준공일_datetime'].dt.year
df['준공일_월'] = df['준공일_datetime'].dt.month
df['준공일_일'] = df['준공일_datetime'].dt.day
```

```python
date1 = datetime.datetime(2020,12,31,0)
d_day= list()
```

```python
for index, row in df.iterrows(): 
    date2 = datetime.datetime(row['가동개시일_연도'],row['가동개시일_월'],row['가동개시일_일'],0)
    td = date1 - date2
    d_day.append(td.days)
    print(td.days)
```

```python
c_day= list()
for index, row in df.iterrows(): 
    date2 = datetime.datetime(row['준공일_연도'],row['준공일_월'],row['준공일_일'],0)
    cd = date1 - date2
    c_day.append(cd.days)
```

```python
df = df.assign(cday = c_day)
```

```python
df = df.assign(dday = d_day)
```

```python
df=df.rename(columns={'dday':'가동개시일_day'})
```

```python
df = df.rename(columns={'cday':'준공일_day'})
```
